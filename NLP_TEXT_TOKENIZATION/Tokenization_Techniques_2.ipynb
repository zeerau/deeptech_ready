## 1. Word-Level Tokenization

```python
import nltk
from nltk.tokenize import word_tokenize

# Install NLTK resources if not already done
nltk.download('punkt')

# Example text
text = "Tokenization is crucial for NLP models!"

# Word-level tokenization
tokens = word_tokenize(text)
print("Word-Level Tokens:", tokens)
```

## 2. Character-Level Tokenization

```python
# Example text
text = "Tokenization is crucial for NLP models!"

# Character-level tokenization
tokens = list(text)
print("Character-Level Tokens:", tokens)
```

## 3. Subword-Level Tokenization
### 3.1 Byte Pair Encoding (BPE)

```python
from tokenizers import Tokenizer

# Load a pre-trained BPE tokenizer
tokenizer = Tokenizer.from_pretrained("gpt2")  # GPT-2 uses BPE

# Tokenize a sample text
output = tokenizer.encode("Tokenization is crucial for NLP models.")
print("BPE Tokens:", output.tokens)
```

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer

# Initialize tokenizer and trainer
tokenizer = Tokenizer(BPE())
trainer = BpeTrainer(special_tokens=["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"])

# Training corpus
corpus = ["Natural Language Processing is fun!", "Machine Learning is amazing!"]

# Train the tokenizer
tokenizer.train_from_iterator(corpus, trainer)

# Tokenize input
bpe_tokens = tokenizer.encode("Natural Language Processing").tokens
print("BPE Tokens:", bpe_tokens)
```

### 3.2 WordPiece Tokenization

```python
from transformers import BertTokenizerFast

# Load a pre-trained WordPiece tokenizer (BERT uses WordPiece)
tokenizer = BertTokenizerFast.from_pretrained("bert-base-uncased")

# Tokenize a sample text
tokens = tokenizer.tokenize("Tokenization is crucial for NLP models.")
print("WordPiece Tokens:", tokens)
```

### 3.3 SentencePiece Tokenization

```python
from transformers import AutoTokenizer

# Load a pre-trained SentencePiece tokenizer (e.g., T5 or ALBERT uses SentencePiece)
tokenizer = AutoTokenizer.from_pretrained("t5-small")

# Tokenize a sample text
tokens = tokenizer.tokenize("Tokenization is crucial for NLP models.")
print("SentencePiece Tokens:", tokens)
```
